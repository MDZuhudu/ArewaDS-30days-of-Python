```python
'''
Python has the module called statistics and we can use this module to do all the statistical calculations. 
However, to learn how to make function and reuse function let us try to develop a program, which calculates 
the measure of central tendency of a sample (mean, median, mode) and measure of variability (range, variance, standard deviation). 
In addition to those measures, find the min, max, count, percentile, and frequency distribution of the sample. You can create a class 
called Statistics and create all the functions that do statistical calculations as methods for the Statistics class. Check the output below.
'''

ages = [31, 26, 34, 37, 27, 26, 32, 32, 26, 27, 27, 24, 32, 33, 27, 25, 26, 38, 37, 31, 34, 24, 33, 29, 26]

import statistics
class Statistics:
    def __init__(self, D):
        self.D = D
    def count(self):
        return len(self.D)
    
    def sum(self):
        return sum(self.D)
    
    def min(self):
        return min(self.D)
    
    def max(self):
        return max(self.D)

    def range(self):
        return self.max() - self.min()
    
    def mean(self):
        return statistics.mean(self.D)

    def median(self):
        return statistics.median(self.D)
    
    def mode(self):
        mode = statistics.mode(self.D)
        count = self.D.count(mode)
        return (mode, count)

    def var(self):
        return statistics.variance(self.D)

    def std(self):
        return statistics.stdev(self.D)

    def freq_dist(self):
        frequency_dict = {}
        for i in self.D:
            if i in frequency_dict:
                frequency_dict[i] += 1
            else:
                frequency_dict[i] = 1
        frequency_list = [(value, key) for key, value in frequency_dict.items()]
        frequency_list.sort(reverse=True)
        return frequency_list

    def describe(self):
        print('Count:', self.count())
        print('Sum: ', self.sum())
        print('Min: ', self.min())
        print('Max: ', self.max())
        print('Range: ', self.range())
        print('Mean: ', self.mean())
        print('Median: ', self.median())
        print('Mode: ', self.mode())
        print('Variance: ', self.var())
        print('Standard Deviation: ', self.std())
        print('Frequency Distribution: ', self.freq_dist())
data = Statistics(ages)
data.describe()
```

```python
'''
Create a class called PersonAccount. It has firstname, lastname, incomes, expenses properties and it has total_income, total_expense,
account_info, add_income, add_expense and account_balance methods. Incomes is a set of incomes and its description. The same goes for expenses.
'''


class PersonAccount:
    def __init__(self, first_name, last_name):
      self.first_name = first_name
      self.last_name = last_name
      self.incomes = []
      self.expense = []
      
    def add_income(self, amount, discription):
     self.incomes.append((amount, discription))
     
    def add_expense(self, amount, discription):
      self.expense.append((amount, discription))
      
    def total_income(self):
      total = 0	
      for income in self.incomes:
        total += income[0]
      return total
      
	  
      
    def total_expense(self):
      total = 0
      for expense in self.expense:
        total += expense[0]
      return total
      
 
    def account_balance(self):
      return self.total_income() - self.total_expense()
    
    def acount_info(self):
      print(f"Full_name: {self.first_name}  {self.last_name}")
      print("Incomes: ", self.incomes)
      print("Expenses: ", self.expense)
      print("Total Income: ", self.total_income())
      print("Total Expense: ", self.total_expense())
      print("Account Balance: ", self.account_balance())
      
account = PersonAccount('Muhammad', 'Dahiru' )
account.add_income(100000, '1st Income')
account.add_income(200000, '2nd Income')

account.add_expense(50000, 'Maintenance')
account.add_expense(12000, 'Diesel')

account.acount_info()
```
